```python
import uproot
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
from ipyfilechooser import FileChooser


root_file_path = r"C:\Users\nika\OneDrive\Dokumenty\python_practical_2026\Project-116\00385292_00000001_1.dvntuple.root"  
file = uproot.open(root_file_path)

print(file.keys())

keys = file.keys()[1]   
tree = file[keys]
print(tree.keys())

#reading branches from the list
all_branches = tree.keys()
variables = list(tree.keys())

###Conditions###
#Add new functions and limits here then change only the combine_masks function to add new cuts to the final plot
#rest should be unchanged


def load_branch(branch_name):
    return tree[branch_name].array(library="np")

def momentum_mask(pt_min=1000):
    pt = load_branch("B0_PT")
    return np.isfinite(pt) & (pt > pt_min)


def mcorr_mask(mcr_min=1554, mcr_max=5743):
    mcr = load_branch("B0_MCORR")
    return (
        np.isfinite(mcr)
        & (mcr > mcr_min)
        & (mcr < mcr_max)
    )


def b0_ipchi2_ownpv(ipchi_max=5):
    ipchi2 = load_branch("B0_IPCHI2_OWNPV")
    return np.isfinite(ipchi2) & (ipchi2 < ipchi_max)


def kaon_probability_mask(k_threshold=0.85):
    kplus = load_branch("Kplus_ProbNNk")
    return np.isfinite(kplus) & (kplus > k_threshold)


def piminus_prob_k(pi_k_max=0.15):
    pi_k = load_branch("piminus_ProbNNk")
    return np.isfinite(pi_k) & (pi_k < pi_k_max)


def kplus_probmu(kp_mu_max=0.15):
    kp_mu = load_branch("Kplus_ProbNNmu")
    return np.isfinite(kp_mu) & (kp_mu < kp_mu_max)


def kp_prob_e(kp_e_max=0.15):
    kp_e = load_branch("Kplus_ProbNNe")
    return np.isfinite(kp_e) & (kp_e < kp_e_max)


def muplus_prob_e(muplus_e_max=0.15):
    muplus_e = load_branch("muplus_ProbNNe")
    return np.isfinite(muplus_e) & (muplus_e < muplus_e_max)


def muminus_prob_e(muminus_e_max=0.15):
    muminus_e = load_branch("muminus_ProbNNe")
    return np.isfinite(muminus_e) & (muminus_e < muminus_e_max)


def kp_prob_p(kp_p_max=0.15):
    kp_p = load_branch("Kplus_ProbNNp")
    return np.isfinite(kp_p) & (kp_p < kp_p_max)


def muplus_prob_p(muplus_p_max=0.15):
    muplus_p = load_branch("muplus_ProbNNp")
    return np.isfinite(muplus_p) & (muplus_p < muplus_p_max)


def muminus_prob_p(muminus_p_max=0.15):
    muminus_p = load_branch("muminus_ProbNNp")
    return np.isfinite(muminus_p) & (muminus_p < muminus_p_max)


def kp_prob_pi(kp_pi_max=0.15):
    kp_pi = load_branch("Kplus_ProbNNpi")
    return np.isfinite(kp_pi) & (kp_pi < kp_pi_max)


def muplus_prob_pi(muplus_pi_max=0.15):
    muplus_pi = load_branch("muplus_ProbNNpi")
    return np.isfinite(muplus_pi) & (muplus_pi < muplus_pi_max)


def muminus_prob_pi(muminus_pi_max=0.15):
    muminus_pi = load_branch("muminus_ProbNNpi")
    return np.isfinite(muminus_pi) & (muminus_pi < muminus_pi_max)


def kplus_not_muon():
    k_mu = load_branch("Kplus_isMuon")
    return np.isfinite(k_mu) & (k_mu == 0)


def piminus_not_muon():
    pi_mu = load_branch("piminus_isMuon")
    return np.isfinite(pi_mu) & (pi_mu == 0)


def muons_are_muons():
    muplus = load_branch("muplus_isMuon")
    muminus = load_branch("muminus_isMuon")

    return (
        np.isfinite(muplus)
        & np.isfinite(muminus)
        & (muplus == 1)
        & (muminus == 1)
    )


def tos_mask():
    tos = load_branch("B0_Hlt2Phys_TOS")
    return np.isfinite(tos) & (tos == 1)


def b0_mass_window(low=5000, high=5500):
    mass = load_branch("B0_MM")
    return np.isfinite(mass) & (mass > low) & (mass < high)


#  Combine all cuts


def combine_masks(
    k_threshold=0.85,
    pt_min=1000,
    mcr_min=1554,
    mcr_max=5743,
    ipchi_max=5,
    pi_k_max=0.15,
    kp_mu_max=0.15,
    kp_e_max=0.15,
    muplus_e_max=0.15,
    muminus_e_max=0.15,
    kp_p_max=0.15,
    muplus_p_max=0.15,
    muminus_p_max=0.15,
    kp_pi_max=0.15,
    muplus_pi_max=0.15,
    muminus_pi_max=0.15,
):
    mask = (
        momentum_mask(pt_min)
        & mcorr_mask(mcr_min, mcr_max)
        & b0_ipchi2_ownpv(ipchi_max)
        & kaon_probability_mask(k_threshold)
        & piminus_prob_k(pi_k_max)
        & kplus_probmu(kp_mu_max)
        & kp_prob_e(kp_e_max)
        & muplus_prob_e(muplus_e_max)
        & muminus_prob_e(muminus_e_max)
        & kp_prob_p(kp_p_max)
        & muplus_prob_p(muplus_p_max)
        & muminus_prob_p(muminus_p_max)
        & kp_prob_pi(kp_pi_max)
        & muplus_prob_pi(muplus_pi_max)
        & muminus_prob_pi(muminus_pi_max)
        & kplus_not_muon()
        & piminus_not_muon()
        & muons_are_muons()
        & tos_mask()
        & b0_mass_window()
    )

    return mask
###Plotting two graphs next to each other
def plot_with_cuts(
    variable,
    k_threshold=0.85,
    pt_min=1000,
    bins=100,
    x_min=None,
    x_max=None
):
    data = load_branch(variable)

    mask = combine_masks(
        k_threshold=k_threshold,
        pt_min=pt_min
    )

    filtered_data = data[mask]

    data = data[np.isfinite(data)]
    filtered_data = filtered_data[np.isfinite(filtered_data)]

    if variable == "B0_PT":
        data = data / 1000
        filtered_data = filtered_data / 1000
        xlabel = "B0_PT [GeV/c]"
    else:
        xlabel = variable

    fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

    axes[0].hist(
        data,
        bins=bins,
        range=(x_min, x_max),
        histtype="step",
        linewidth=1.5
    )
    axes[0].set_title("Original")
    axes[0].set_xlabel(xlabel)
    axes[0].set_ylabel("Number of events")

    axes[1].hist(
        filtered_data,
        bins=bins,
        range=(x_min, x_max),
        histtype="step",
        linewidth=1.5
    )
    axes[1].set_title("After cuts")
    axes[1].set_xlabel(xlabel)

    plt.tight_layout()
    plt.show()

    print("Original events:", len(data))
    print("Events after cuts:", len(filtered_data))
    print("Removed events:", len(data) - len(filtered_data))

# Interactive part- you can add sliders if we are not sure aboutr constants


variable_dropdown = widgets.Dropdown(
    options=variables,
    description="Variable:"
)

bins_slider = widgets.IntSlider(
    value=100,
    min=1,
    max=300,
    step=1,
    description="Bins:"
)

k_threshold_slider = widgets.FloatSlider(
    value=0.85,
    min=0,
    max=1,
    step=0.01,
    description="K+ prob:"
)

pt_min_slider = widgets.FloatSlider(
    value=1000,
    min=0,
    max=10000,
    step=100,
    description="B0 PT min:"
)

widgets.interact(
    plot_with_cuts,
    variable=variable_dropdown,
    k_threshold=k_threshold_slider,
    pt_min=pt_min_slider,
    bins=bins_slider,
    x_min=widgets.FloatText(value=0, description="x min"),
    x_max=widgets.FloatText(value=6000, description="x max")
)
```
